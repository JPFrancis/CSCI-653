```python
import os

timeStepFiles = []
for root, dirs, files in os.walk("./Graphene"):
    for file in files:
        if file.endswith(".txt"):
            timeStepFiles.append(file)

print(len(timeStepFiles))
```

```python
import numpy as np

coordinates = np.zeros((201, 4358, 3))
maxX = 0
maxY = 0
maxZ = 0
minX = 0
minY=  0
minZ = 0

for t, timeStepFile in enumerate(timeStepFiles):
    f=open("./Graphene/" + timeStepFile, "r")
    contents =f.readlines()

    atomV = []
    for i in range(2, len(contents)):
        values = contents[i].split("    ")
        atomV.append(values[0])
        coordinates[t][i-2][0] = float(values[1])
        coordinates[t][i-2][1] = float(values[2])
        coordinates[t][i-2][2] = float(values[3][:-2])
   
print(np.max(coordinates[:,:,0]))
print(np.max(coordinates[:,:,1]))
print(np.max(coordinates[:,:,2]))
```

```python
image = np.zeros((201, 256, 256, 256))

for i in range(201):
    for coordinate in coordinates[i]:
        image[i][int(coordinate[0])][int(coordinate[1])][int(coordinate[2])] = 255.0
```
